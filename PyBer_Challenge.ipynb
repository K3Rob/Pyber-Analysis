# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
pyber_data_df["date"] = pd.to_datetime(pyber_data_df["date"])
pyber_data_df.dtypes
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

rides_per_type = pyber_data_df.groupby('type').count()['ride_id']
rides_per_type
```

```python
# 2. Get the total drivers for each city type

drivers_per_type = city_data_df.groupby('type').sum()['driver_count']
drivers_per_type
```

```python
#  3. Get the total amount of fares for each city type
fares_per_type = pyber_data_df.groupby('type').sum()['fare']
fares_per_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = fares_per_type / rides_per_type
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = fares_per_type / drivers_per_type
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    'Total Rides': rides_per_type,
    'Total Drivers': drivers_per_type,
    'Total Fares': fares_per_type,
    'Average Fare per Ride': avg_fare_per_ride,
    'Average Fare per Driver': avg_fare_per_driver})

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.

pyber_summary_df = pd.DataFrame({
    'Total Rides': rides_per_type.map("{:,}".format),
    'Total Drivers': drivers_per_type.map("{:,}".format),
    'Total Fares': fares_per_type.map("${:,.2f}".format),
    'Average Fare per Ride': avg_fare_per_ride.map("${:,.2f}".format),
    'Average Fare per Driver': avg_fare_per_driver.map("${:,.2f}".format)})

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_table_df = pyber_data_df.groupby(['type','date']).sum()['fare']
pyber_table_df
```

```python
# 2.  Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_table_df =pyber_table_df.reset_index()
pyber_table_df.head(10)
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_table_df = pyber_table_df.pivot("date","type", "fare")
pyber_table_df
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
#pyber_date_df = pyber_table_df.loc[pyber_table_df['date'].between('2019-01-01','2019-04-29')]
pyber_date_df = pyber_table_df.loc['2019-01-01':'2019-04-29']
pyber_date_df
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

pyber_date_df.index = pd.to_datetime(pyber_date_df.index)

pyber_date_df
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
pyber_date_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_by_week_df = pyber_date_df.resample('W').sum()

pyber_by_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

pyber_by_week = pyber_by_week_df.plot()
plt.title("Total Fare by City Type")
plt.ylabel('Fare ($USD)')
plt.xlabel(' ')
#ax.set_xlim(2019-1-6, 2019-4-28)

# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.
plt.style.use('fivethirtyeight')

```
